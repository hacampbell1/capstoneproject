# Business Statement

- Problem solved
    - Classification algorithm to identify fraud bank account applications

- If we resample before train-test split the split distribution may not replicate the distribution of real-world data
- We resample to increase the impact of the minority class on the cost function
- False Negative: Truth is a fraud application (1) but model predicts non-fraud (0)
- False Positive: Truth is a non-fraud application (0) but the model predicts fraud (1)
- In our case false negative is way more important, thus we are not only minimizing the overall misclassification rate, but prioritize minimzing the false negative rate
- SMOTE generate noisy samples by interpolating new points between marginal outliers and inliers. This issue can be solved by cleaning the space resulting from over-sampling. We use SMOTEENN

# Define Functions 

```python
def target_class_proportion(y_train, y_test):
    print('Proportion of Fraud vs. Non-Fraud in Training: ')
    print(y_train['fraud_bool'].value_counts()/len(y_train))
    print('\n')
    print('Proportion of Fraud vs. Non-Fraud in Test: ')
    print(y_test['fraud_bool'].value_counts()/len(y_test))

    
    
def target_class_counts(y_train, y_resampled, resampling):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5), dpi = 150)

    
    sns.countplot(x = y_train['fraud_bool'], ax = ax1)
    ax1.set_title('Class Distribution in Train Before {}'.format(resampling))
    ax1.set_xlabel('Non-Fraud (0) vs. Fraud (1)')
    sns.countplot(x = y_resampled['fraud_bool'], ax = ax2)
    ax2.set_title('Class Distribution in Train After {}'.format(resampling))
    ax2.set_xlabel('Non-Fraud (0) vs. Fraud (1)')

    plt.show()
    

    
def visualize_distribution(feature):
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))

    sns.histplot(baf_base[feature], kde = True, ax = ax1)
    sns.boxplot(baf_base[feature], ax = ax2)

    ax1.set_xlabel(feature)
    ax1.set_title('Histogram Plotted for {}'.format(feature))
    
    ax2.set_xlabel(feature)
    ax2.set_title('Boxplot Plotted for {}'.format(feature))
    plt.show()

def histogram_scatterplot(X, y, feature):
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))
    
    sns.histplot(X[feature], kde = True, ax = ax1)
    sns.scatterplot(x = X[feature], y = y['fraud_bool'], ax = ax2)

    ax1.set_xlabel(feature)
    ax1.set_title('Histogram Plotted for {}'.format(feature))
    
    ax2.set_xlabel(feature)
    ax2.set_title('Scatterplot Plotted for {}'.format(feature))
    plt.show()

def compare_train_test_distribution(train, test, feature):
    plt.figure(figsize=(12, 6))
    sns.kdeplot(data = train, x = feature, label='Train', fill=True)
    sns.kdeplot(data = test, x = feature, label='Test', fill=True)
    plt.title(f'Distribution Comparison of customer_age between Train and Test Sets')
    plt.legend()
    plt.show()
    
    
# Check Binary Instances

def compare_binary_dist(feature):
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))
    sns.barplot(x = X_train[feature].value_counts().index, y = X_train[feature].value_counts(), ax = ax1)
    sns.barplot(x = X_train_processed[feature].value_counts().index, y = X_train_processed[feature].value_counts(), ax = ax2)

#SMOTE changed the distribution of binary variables
    
def get_template_pr_curve(ax, title,fs, baseline=0.5):
    ax.set_title(title, fontsize=fs)
    ax.set_xlim([-0.01, 1.01])
    ax.set_ylim([-0.01, 1.01])
    
    ax.set_xlabel('Recall (True Positive Rate)', fontsize=fs)
    ax.set_ylabel('Precision', fontsize=fs)
    
    ax.plot([0, 1], [baseline, baseline],'r--',label='AP Random = {0:0.3f}'.format(baseline))
    
def classification_metrics(y, y_pred, model):
    
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred)
    recall = recall_score(y, y_pred)
    f1 = f1_score(y, y_pred)
    roc_auc = roc_auc_score(y, y_pred)
    
    print(f'Accuracy: {accuracy:.2f}')
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1 Score: {f1:.2f}')
    print(f'ROC AUC: {roc_auc:.2f}')

    print('Confusion Matrix:')


    from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

    cnf_matrix = confusion_matrix(y, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cnf_matrix, display_labels=model.classes_)
    disp.plot(cmap=plt.cm.Blues)

#Functions to evaluate threshold based & threshold free metrics

def get_class_from_fraud_probability(fraud_probabilities, threshold=0.5):
    
    predicted_classes = [0 if fraud_probability<threshold else 1 
                         for fraud_probability in fraud_probabilities]

    return predicted_classes


def threshold_based_metrics(fraud_probabilities, true_label, thresholds_list):
    
    results = []
    
    for threshold in thresholds_list:
    
        predicted_classes = get_class_from_fraud_probability(fraud_probabilities, threshold=threshold)
    
        (TN, FP, FN, TP) = confusion_matrix(true_label, predicted_classes).ravel()
    
        MME = (FP+FN)/(TN+FP+FN+TP)
    
        TPR = TP/(TP+FN)
        TNR = TN/(TN+FP)
    
        FPR = FP/(TN+FP)
        FNR = FN/(TP+FN)
        
        BER = 1/2*(FPR+FNR)
        
        Gmean = np.sqrt(TPR*TNR)
    
        precision = 0
        FDR = 0
        F1_score=0
        
        if TP+FP>0:
            precision = TP/(TP+FP)
            FDR=FP/(TP+FP)
        
        NPV = 0
        FOR = 0
        
        if TN+FN>0:
            NPV = TN/(TN+FN)
            FOR = FN/(TN+FN)
            
        
        if precision+TPR>0:
            F1_score = 2*(precision*TPR)/(precision+TPR)
    
        results.append([threshold, MME, TPR, TNR, FPR, FNR, BER, Gmean, precision, NPV, FDR, FOR, F1_score])
        
    results_df = pd.DataFrame(results,columns=['Threshold' ,'MME', 'TPR', 'TNR', 'FPR', 'FNR', 'BER', 'G-mean', 'Precision', 'NPV', 'FDR', 'FOR', 'F1 Score'])
    
    return results_df

def evaluate(y_pred, FIXED_FPR):
    
    fprs, tprs, thresholds = roc_curve(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    plt.figure(figsize=(12, 8))
    sns.lineplot(fprs, tprs, label=f'AUC = {roc_auc:.2f}')
    plt.plot([0, 1], [0, 1], 'k--', label='Random Guessing')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()
    
    tpr = tprs[fprs<FIXED_FPR][-1]
    fpr = fprs[fprs<FIXED_FPR][-1]
    threshold = thresholds[fprs<FIXED_FPR][-1]
        
    print("AUC:", roc_auc_score(y_test, y_pred))
    to_pct = lambda x: str(round(x, 4) * 100) + "%"
    print("TPR: ", to_pct(tpr), "\nFPR: ", to_pct(fpr), "\nThreshold: ", round(threshold, 2))
    
def model_fit_predict(classifier, X_train, y_train, X_test, y_test):
    
    classifier.fit(X_train, y_train)

    y_train_pred = classifier.predict(X_train)
    y_test_pred = classifier.predict(X_test)
    
    y_train_pred_prob = classifier.predict_proba(X_train)[:, 1]
    y_test_pred_prob = classifier.predict_proba(X_test)[:, 1]
    
    return y_train_pred, y_test_pred, y_train_pred_prob, y_test_pred_prob


def plot_learning_curves(estimator, X, y, title, ylim=None, cv=None,
                         n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate learning curves for training and testing datasets.

    Parameters:
    - estimator: the machine learning model
    - X: feature matrix
    - y: target variable
    - title: title of the plot
    - ylim: tuple (min, max), optional - defines minimum and maximum y-axis values
    - cv: cross-validation iterator, optional - default is 5-fold cross-validation
    - n_jobs: number of jobs to run in parallel, optional
    - train_sizes: array-like, shape (n_ticks,), optional - Relative or absolute numbers of training examples
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Precision")

    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring='precision')

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training precision")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Testing precision")

    plt.legend(loc="best")
    return plt

def rf_feature_importances(X_train, classifier):

    importances = rf_classifier.feature_importances_

    # Get the feature names
    feature_names = X_train.columns

    # Create a DataFrame for easier plotting
    df_importances = pd.DataFrame({'Feature': feature_names, 'Importance': importances})

    # Sort the DataFrame by importance in descending order
    df_importances = df_importances.sort_values(by='Importance', ascending=False).head(10)

    # Plot the feature importances vertically using sns.barplot
    plt.figure(figsize=(10, 8), dpi = 200)
    plt.barh(df_importances['Feature'], df_importances['Importance'], data=df_importances, height = 0.5, edgecolor='black', linewidth=1)
    plt.xlabel('Feature Importance')
    plt.xticks([])
    plt.ylabel('Feature')
    plt.yticks(fontsize=14)
    plt.title('Top Random Forest Features')
    plt.gca().invert_yaxis()
    plt.show()
    
def xgb_feature_importances(classifier):
    importances = xgb_classifier.get_booster().get_score(importance_type="gain")

    sorted_importances = sorted(importances.items(), key=lambda x: x[1], reverse=True)

    features, gain_values = zip(*sorted_importances)

    top_features = features[:10]
    top_gain_values = gain_values[:10]

    # Plot the feature importances vertically using sns.barplot
    plt.figure(figsize=(10, 8), dpi = 200)
    plt.barh(top_features, top_gain_values, height = 0.5, color = 'orangered', edgecolor='black', linewidth=1)
    plt.xlabel('Feature Gain')
    plt.xticks([])
    plt.ylabel('Feature')
    plt.yticks(fontsize=14)
    plt.title('Top XGBoost Features')
    plt.gca().invert_yaxis()
    plt.show()
```
